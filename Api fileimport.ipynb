```python
import pandas as pd
import requests
import json
```

```python
# for example code use for get convert api data to dataframe

url = "https://exampleapi.p.rapidapi.com/data"

headers = {
    "X-RapidAPI-Key": "YOUR_API_KEY",
    "X-RapidAPI-Host": "exampleapi.p.rapidapi.com"
}

response = requests.get(url, headers=headers)
data = response.json()

# Check data structure
print(data)

# Convert to DataFrame
df = pd.json_normalize(data)

print(df.head())
```

for see json data in downloadlble file ,must import json before use

```python
with open('data.json', 'w') as f:
    json.dump(data, f, indent=4)
```
